## 2. 크롤링 작업

- 간혹가다가 크롤링이 잘 안되는 페이지들이 있을 수 있다! 

- 한글이 깨질 경우, 'request' 대신 'urllib.request'를 이용해본다.

```
url = 'https://m.etnews.com/20200831000113'
source_code_from_url = urllib.request.urlopen(url)
soup = BeautifulSoup(source_code_from_url, 'html.parser')
print(soup)
```

```
body = soup.find('div',{'class':'article-body'})
body
```

```
p = body.find_all('p')
p
```

```
article=''

for link in p:
  article+= link.text

print(article)
```

```
from konlpy.tag import Mecab
m = Mecab()
m.nouns('메켑 설치가 되었는지 확인')
```

```
mecab = Mecab()
mecab_tokens = mecab.morphs(article)

print(type(mecab_tokens))
print(mecab_tokens[:20])
```

## 4. 품사 부착해주기

```
mecabTag = []

for token in mecab_tokens:
  mecabTag += mecab.pos(token)

print(mecabTag[:20])
```

## 5. 불용어 처리하기
- 최빈어 조회 먼저 한 후, 불용어 처리를 진행한다.

```
from collections import Counter
Counter(mecabTag).most_common()
```

```
print(mecab_tokens)
```

```
stopPos = ['Determiner','Adverb','Conjunction','Josa','PreEomi','Eomi',
           'Suffix','Punctuation','Foreign','Alpha','Number','Unknown']

stopword = ['이','에','하','는','을','해','는','도']

word =[]
for tag in mecabTag:
  if tag[1] not in stopPos:
    if tag[0] not in stopword:
      word.append(tag[0])

print(word)
```
