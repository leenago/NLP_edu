```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.forbes.com/sites/adrianbridgwater/2019/04/15/what-drove-the-ai-renaissance/#2302bc011f25'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser') #res 안에 있는 내용을 text처리 하겠다
print(soup)
```

```python
print(soup.prettify())  #soup으로 된 내용을 조금 더 보기 좋게 정리
```

```python
body = soup.find("div", {'class':'article-body fs-article fs-responsive-text current-article'})
body #div 태그에 class 이름이 article~ 인 부분을 가져와라

#body = soup.find('div', class_='article~')
#이렇게 작성하면 class 뒤에 언더바 필수!
```

```python
p = body.find_all('p') #div 안에 있는 전체p를 찾아내는 과정, list로 받아진다
p
```

```python
article=''

for link in p:
    article += link.text #이렇게 해주어야 했군,,!
print(article)
    
#article에 link.text를 더해서 그 값을 다시 article에 넣어주란 뜻
```

## 영문토큰화

#### word_tokenize(): 마침표와 구두점으로 구분하여 토큰화

```python
!pip install nltk
```

```python
import nltk

nltk.download('punkt') #tokenizer를 하려면 품사부착 등의 기능을 수행해주기 위해서 필요한 데이터 뭉치
#한번 적으면 이후에는 적지 않아도 괜찮다.
from nltk.tokenize import word_tokenize
token1 = word_tokenize(article)
print(token1[:20])
```

#### WordPunctTokenizer(): 알파벳이 아닌 문자를 구분하여 토큰화

```python
import nltk
from nltk.tokenize import WordPunctTokenizer
token2 = WordPunctTokenizer().tokenize(article)
print(token2[:20])
```

#### TreebankWordTokenizer(): 정규표현식에 기반한 토큰화

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
token3 = TreebankWordTokenizer().tokenize(article)
print(token3[:20])
```

## 영문 품사부착
#### 분리한 토큰마다 품사를 부착한다
#### 태그 목록: https://pythonprogramming.net/natural-language-toolkit-nltk-part-speech-tagging/

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
```

```python
taggedToken = pos_tag(token1) #pos_tag에 토큰화된 개체를 집어넣기
print(taggedToken[:20])
```

## 영문 개체명인식

```python
nltk.download('words')
nltk.download('maxent_ne_chunker')

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

#토큰화
token1 = word_tokenize('Barak Obama likes fried chicken very much')
print('token: ', token1)

#pos-tag
taggedToken = pos_tag(token1)
print('pos-tag: ', taggedToken)

#chunking
from nltk import ne_chunk
neToken= ne_chunk(taggedToken)
print('ne-token: ', neToken)
```
